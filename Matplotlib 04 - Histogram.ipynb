```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Data

```python
mu, sigma = 100, 15 # mean and standard deviation

x = mu + sigma * np.random.randn(10000) # normal distribution
```

```python
x.shape
```

## Pyplot Style

```python
plt.hist(x, bins=50, facecolor='g', alpha=0.5)

plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Contoh Histogram')

plt.text (45, 500, '$\mu=100, \ \sigma=15$') #Parameter (coordinate, coordinate, text)
plt.grid()

plt.show()
```

## OO Style

```python
fig, ax = plt.subplots()

ax.hist(x, bins=50, facecolor='g', alpha=0.5)

ax.set_xlabel('Sumbu x')
ax.set_ylabel('Sumbu y')
ax.set_title('Contoh Histogram')

ax.text (45, 500, '$\mu=100, \ \sigma=15$') #Parameter (coordinate, coordinate, text)
ax.grid()

plt.show()
```
