```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
# Vertical bar plot
plt.bar(kategori, data,
       color='red', # Color
       alpha=0.25)  # Transparansi
plt.grid(linestyle='--',
        linewidth=2,
        axis='both',   # Choice = x, y, both
        alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
# Horizontal Bar plot
plt.barh(kategori, data,
       color='red', # Color
       alpha=0.25)  # Transparansi

plt.grid(linestyle='--',
        linewidth=2,
        axis='both',   # Choice = x, y, both
        alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
# Horizontal Bar plot
plt.barh(kategori[::-1], data[::-1], # Change list of the data
       color='red', # Color
       alpha=0.25)  # Transparansi

plt.grid(linestyle='--',
        linewidth=2,
        axis='both',   # Choice = x, y, both
        alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## Stacked Bar plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label = 'Data 1')
plt.bar(kategori, data2, label = 'Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
        linewidth=2,
        axis='y',   # Choice = x, y, both
        alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## Grouped Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
```

```python
width = 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori) # Creating kategori on X axis

plt.grid(linestyle='--',
        linewidth=2,
        axis='y',   # Choice = x, y, both
        alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.legend()
plt.show()
```
