# There is 2 ways to use Matplotlib

- OO Style( Object Oriented style)
- Pyplot style

## OO Style

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots() # Creating a figure and axes

# Plotting 3 variant data on axes
ax.plot (x, x, label='linear')
ax.plot (x, x**2, label='Quadratic')
ax.plot (x, x**3, label='Cubic')

ax.set_xlabel('x label') # Creating x-label on axes
ax.set_ylabel('y label') # Creating y-label on axes
ax.set_title('Simple Plot') # Creating title on axex
ax.legend() # Show Legend
```

## Pyplot Style

```python
x = np.linspace(0, 2, 100)
x
```

```python
# Plotting 3 variant data on axes
plt.plot (x, x, label='linear')
plt.plot (x, x**2, label='Quadratic')
plt.plot (x, x**3, label='Cubic')

plt.xlabel('x label') # Creating x-label on axes
plt.ylabel('y label') # Creating y-label on axes
plt.title('Simple Plot') # Creating title on axex
plt.legend() # Show Legend
```
