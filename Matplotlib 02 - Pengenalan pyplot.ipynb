```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Creating simple plotting

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
```

## Setting format plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go')
plt.axis([0, 6, 0, 20])
plt.show()
```

- the third parameter on the plt.plot is parameter for creating color and type formatting
- Function axis() is represent of list [xmin, xmax, ymin, ymax]

## Multiplot 

```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')

plt.show()
```

### Alternative multiplot

```python
plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^')

plt.show()
```

## Plotting with keywords

```python
data = {'a' : np.arange(50),
       'c' : np.random.randint(0, 50, 50),
       'd' : np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data= data)

plt.xlabel('entry a')
plt.ylabel('entry b')

plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
# Parameter c = color and s = size

plt.xlabel('entry a')
plt.ylabel('entry b')

plt.show()
```
